# k-means 

Imagine una situación en la que una empresa de supermercados quiere abrir una sucursal de entrega en línea. La empresa existe desde hace mucho tiempo en el mercado y, a través de las tarjetas de membresía, tienen toda la información sobre los clientes, como: cédula del cliente (se puede vincular a toda la información del cliente), productos favoritos, edad, sexo, ingresos anuales, puntuación de gastos, dirección etc. (Exactamente por eso siempre te piden tu cédula al hacer la compra - para organizar mejor la empresa y hacerla más eficiente). Con base en toda esta información, la empresa intentará configurar las revistas en línea de tal manera que puedan cubrir la mayor parte del área y tener productos que satisfagan las necesidades de los clientes. La herramienta que es perfectamente adecuada para esta tarea es "K-Means Clustering". Con el análisis a la empresa le gustaría saber por ejemplo:

* qué grupos de clientes tiene el supermercado
* cuántos centros de distribución pueden abrir en una ubicación determinada
* ¿Cuáles son los productos más importantes que el grupo objetivo ordenará primero, etc.?

El problema es multidimensional porque hay muchas características que caracterizan al cliente. Pero la idea básica del algoritmo k-means se puede visualizar con solo dos características.

```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs
```

```python
np.random.seed(2)
```

Produzcamos artificialmente algunos datos utilizando una biblioteca especial de sklearn llamada `blobs`.
* ganamos 5000 puntos (`n_samples=`)
* con cuatro centros diferentes (especificados como parámetros, `centers=`)
* y esos grupos tendrán una cierta extensión (`cluster_std=`)

```python
X, Y = make_blobs(n_samples=5000, centers=[[40,48], [80, 90], [65, 50], [40, 82]], cluster_std=10.0)
```

```python
fig,ax=plt.subplots(1,1)
ax.scatter(X[:, 0], X[:, 1], marker='.')
ax.set_title('customer segmentation')
ax.set_xlabel('annual income (1000 USD/year)')
ax.set_ylabel('spending score')
```

Generamos cuatro grupos distintos de "clientes" y podemos verlos a simple vista. Pero, ¿cómo encuentra la computadora esos grupos? Recuerde que en este ejemplo observamos solo dos características: 2D, pero las dimensiones superiores no son tan intuitivas). Así que necesitamos un algoritmo.

El algoritmo k-means es un algoritmo iterativo que intenta dividir el conjunto de datos en k subgrupos distintos no superpuestos predefinidos (clusters) donde cada punto de datos pertenece a un solo grupo (cluster).

El algoritmo intentará:

* mantenga los puntos de datos dentro del grupo lo más similares posible y al mismo tiempo mantenga los grupos diferentes tan lejos como sea posible
* asignar puntos de datos a un grupo de manera que la suma de la distancia al cuadrado entre los puntos de datos y el centroide del grupo (media aritmética de todos los puntos de datos que pertenecen a ese grupo) sea mínima.
* realizar la pequeña variación dentro de los clusters. Cuanto más homogéneos (similares) sean los puntos de datos dentro del mismo grupo sera mejor

La operación del algoritmo se puede dividir en cuatro etapas. El único parámetro inicial que debemos proporcionar es escoger numero de grupos en que queremos dividir los datos (clusters), así que elegimos el número de centroides. El centroide es un punto especial que se asignará a un grupo particular de puntos (si elegimos cuatro centroides, cada centroide se asignará a sí mismo un grupo de puntos: cuatro grupos de puntos). Despues de esta eleccion el algoritmo hace lo siguiente:

1. Coloque aleatoriamente k centroides (coordinadas aleatorias), uno para cada grupo.
2. Calcular la distancia de cada punto a cada centroide (distancia euclidiana entre cada coordenada de punto y cada centroide)
3. Asigne cada punto de datos a su centroide más cercano (creando un grupo).
4. Recalcular las posiciones de los k centroides (la nueva posición del centroide corresponderá a la posición media de los puntos asignados al centroide)
5. Repita los pasos 2–4 hasta que todos los centroides ya no se muevan

Es más claro si visualizamos el proceso con la siguiente animación del algoritmo k-means:

* las estrellas son centroides (en la animación son 3 centroides)
* durante iteraciones los centroids se mueven y se agrupan diferentes subconjuntos de los puntos (mas cerca)
* los centroides encuentran un equilibrio y el algoritmo termina

![](imgs/animation.gif)

Realmente esto es tambien un problema de tipo minimización! La función objetivo en este caso se puede expresar con la siguiente expresión:

![](imgs/minimization_function_I.png)

* la suma interna es la suma de las distancias de cada punto a su centroide ($x_i$ es un punto, $c_j$ es un centroide)
* la suma externa es la suma sobre el número del centroides (`k`)
* por lo tanto, el objetivo es minimizar la suma de todas las distancias entre todas los centroides y sus puntos 

# Implementación

Sklearn tiene el algoritmo `k-means` en su biblioteca y podemos implementarlo rápidamente. Como siempre, primero importamos la biblioteca.

```python
from sklearn.cluster import KMeans 
```

Crear una instancia del modelo pasando los siguientes argumentos
* `init = "k-means++"`: selecciona los centros de clúster iniciales para la agrupación en clústeres k-mean de una manera inteligente (no aleatorio) para acelerar la convergencia
* `n_clusters = 4` - elección del número de clústeres
* `n_init = 12` - número de veces que se ejecutará el algoritmo `k-means` con diferentes semillas de centroide. Los resultados finales serán la mejor salida de `n_init` ejecuciones consecutivas en términos de inercia.

```python
model = KMeans(init = "k-means++", n_clusters = 4, n_init = 12)
```

* entrenando al modelo

```python
model.fit(X)
```

```python
X
```

* en resultado, cada elemento de `X` tiene coordenada de dos características y el algoritmo le asignará un número (como una etiqueta) diciendo a qué grupo de puntos pertenece cada `X`. Como elegimos 4 centroides, habrá cuatro etiquetas: `0,1,2,3`

```python
k_means_labels = model.labels_
k_means_labels
```

* también tenemos acceso a las coordenadas optimizadas de cada uno de los cuatro centroides:

```python
k_means_cluster_centers = model.cluster_centers_
k_means_cluster_centers
```

Teniendo toda esa información, podemos visualizar el efecto de la agrupación de k-means:

```python
fig = plt.figure(figsize=(6, 4))

colors = plt.cm.Spectral(np.linspace(0, 1, 4))

ax = fig.add_subplot(1, 1, 1)

for k, color in zip(range(4), colors):

    group_members = (k_means_labels == k)
    
    cluster_center = k_means_cluster_centers[k]
    
    ax.plot(X[group_members, 0], X[group_members, 1], 'w', markerfacecolor=color, marker='.')
    
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=color,  markeredgecolor='k', markersize=6)

ax.set_title('customer segmentation')
ax.set_xlabel('annual income (kUSD)')
ax.set_ylabel('spending score')
```

dónde:

* `colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means_labels))))` :  podemos asignar a cada grupo de puntos a diferente color (cuatro vectores RGB)
* `for k, color in zip(range(4), colors):` - ciclo de iteración: esto es para darle a cada grupo de puntos un color diferente k será: 0,1,2,3
* `group_members = (k_means_labels == k)` - para cada iteración del ciclo, solo seleccionamos un grupo
* `cluster_center = k_means_cluster_centers[k]` - y cada centroide que pertenece a ese grupo
* `ax.plot(X[group_members, 0], X[group_members, 1], 'w', markerfacecolor=color, marker='.')` - para cada iteración trazamos los puntos seleccionados de `X` (0 - primera característica, 1 - segunda característica)
* y de manera similar trazamos los centroides

Elegimos arbitrariamente cuatro grupos para los puntos dados, pero esto no es científico. De hecho, no lo es, pero hay otra manera de encontrar el número óptimo de los centroides para puntos de datos dados. Este método se llama `método del codo` (`elbow`). Este método ejecutamos el algoritmo muchas veces cada vez cambiando el número de centroides `k` y al final trazamos la variable llamada `inercia vs k`

* la inercia simplemente dice qué tan lejos están los puntos dentro de un grupo. Así que uno quiere lograr la pequeña inercia posible. Pero si elegimos el número de centroides `k` tal que es igual al número de puntos de datos, entonces la inercia es cero. Por lo tanto, debe haber un número óptimo de `k`. Este punto es exactamente el 'punto del codo'.

![](imgs/elbow.png)

Así que vamos a ejecutar el algoritmo 6 veces

```python
def cluster_inertias(n):
    inertias=[]
    kmeans=[]
    outputs=[]
    K=[i for i in range(1,n+1)]    
    for i in range(1,n+1):
        inertia=0
        k_means=KMeans(n_clusters=i,random_state=82)
        k_means.fit(X)
        kmeans.append(k_means)
        inertias.append(k_means.inertia_)
    return inertias,K,n
```

```python
inertias,K,n=cluster_inertias(6)
```

y trazar los resultados:

```python
fig,ax = plt.subplots(1,1)
ax.plot(K,inertias)
ax.set_ylabel("Inertia ( Total Distance )")
ax.set_xlabel("K Value")
ax.grid(True)
```

* Podemos ver que el codo está en k=4, o k=3 (el método del codo es bastante subjetivo). De hecho, la distribución de los puntos de datos en nuestro conjunto de datos se puede dividir igualmente en tres grupos (k = 3). Vamos a ver:

```python
model = KMeans(init = "k-means++", n_clusters = 3, n_init = 10)
model.fit(X)
k_means_labels = model.labels_
k_means_cluster_centers = model.cluster_centers_
```

```python
fig = plt.figure(figsize=(6, 4))

colors = plt.cm.Spectral(np.linspace(0, 1, 3))

ax = fig.add_subplot(1, 1, 1)

for k, color in zip(range(3), colors):

    group_members = (k_means_labels == k)
    
    cluster_center = k_means_cluster_centers[k]
    
    ax.plot(X[group_members, 0], X[group_members, 1], 'w', markerfacecolor=color, marker='.')
    
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=color,  markeredgecolor='k', markersize=6)

ax.set_title('customer segmentation')
ax.set_xlabel('annual income (kUSD)')
ax.set_ylabel('spending score')
```
