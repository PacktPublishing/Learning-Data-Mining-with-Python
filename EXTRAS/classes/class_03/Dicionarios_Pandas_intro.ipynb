## ¿De qué se trata la ciencia de datos?

La ciencia de datos es un enfoque multidisciplinario para extraer información procesable de los grandes volúmenes de datos. 

Los datos se refieren a distintas piezas de información generalmente formateadas y almacenadas de una manera que concuerda con un propósito específico. Los datos pueden existir en varias formas: como números o textos registrados en papel, como bits o bytes almacenados en la memoria electrónica o como hechos que viven en la mente de una persona. Sin embargo, desde el advenimiento de la informática, los datos se refieren más comúnmente a la información que se transmite o almacena electrónicamente.

Las personas que trabajan con los datos se llaman científicos de datos y requieren habilidades informáticas y de ciencia pura para utilizar los datos y resolver diversos problemas relacionados con estos. Un científico de datos debe ser capaz de hacer lo siguiente:

* Aplicar las matemáticas, la estadística y el método científico.

* Utilizar una amplia gama de herramientas y técnicas para evaluar y preparar datos (minería de datos).

* Extraer información de los datos mediante análisis predictivos e inteligencia artificial (IA), incluidos el aprendizaje automático y los modelos de aprendizaje profundo.[23]

* Escribir aplicaciones que automaticen el procesamiento de datos y los cálculos.

* Explicar cómo se pueden utilizar estos resultados para resolver diferentes problemas.

## Tipos de datos
#### Los tipos de datos que suelen ser objeto de análisis y aprendizaje automático son generalmente numéricos y categóricos. 

* Las variables numéricas pueden ser de dos tipos:

  * Discretas: solo tienen valores enteros, por ejemplo, número de cuartos, número de pasajeros.
  * Continuas: tienen cualquier valor, incluidos los números flotantes, por ejemplo, peso, altura.

* Las variables categóricas pueden ser de los siguientes tipos:

  * Nominales: cuando las categorías no se pueden ordenar de ninguna manera, por ejemplo, nombres, colores, sexo.
  * Ordinales: en este caso se pueden ordenar las categorías, por ejemplo: tamaño (pequeño, mediano, grande); clase social (baja, media, alta); nivel de educación (básico, medio, pregrado, posgrado).

Por ejemplo, puede cargarse un texto entero como una cadena:

```python
with open('data/the-zen-of-python.txt','r') as f: # abrir documento para leer (read: 'r') como un alias "f" 
    contents = f.read()                           # leer texto entero como cadena
    print(contents)      
```

```python
type(contents)
```

Ahora se puede, por ejemplo, dividir la cadena por un carácter en particular o líneas nuevas `\n`. (o hacer cualquier método en la cadena). 

```python
contents_lines=contents.split('\n')
```

```python
contents_lines[0]
```

```python
contents_lines[-1]
```

Usando `open` también se puede escribir texto en un archivo:

```python
lista_palabras = ["Uno", "dos", "tres", "cuatro"]
with open('data/documento.txt','w') as f: # abrir documento para escribir (write: 'w') como un alias "f" 
    for palabra in lista_palabras:
        f.write(palabra+'\n')            # '\n' significa nueva línea                    
```

## ¿Qué son los diccionarios?

Diccionario es un objeto muy importante en ciencia de datos y de Python, porque muchos datos para el aprendizaje automático vienen ya en forma de diccionario. 

Diccionario, como su nombre lo indica, contiene dos elementos: "clave" y "valor". Cada clave está separada de su valor por dos puntos (:), los elementos están separados por comas y todo está encerrado en llaves rizadas. Un diccionario vacío, sin ningún elemento, se escribe solo con dos llaves rizadas, como estas: {}. [15]


Un ejemplo de diccionario muy simple puede ser el siguiente:

```python
diccionario_1={1: 'manzana', 2: 'mandarina', 3: 'naranja', 4: 'lemon'}
```

```python
diccionario_1
```

Pueden verse todas claves que tiene este diccionario usando el método `keys()`

```python
diccionario_1.keys()
```

Y ver el valor de cada clave:

```python
diccionario_1[3]
```

También, pueden verse todos los valores que tiene este diccionario usando el método `values()`

```python
diccionario_1.values()
```

Las claves no tienen que ser numéricas y el diccionario anterior se puede "voltear"

```python
diccionario_2={'manzana':1, 'mandarina':2, 'naranja': 3, 'lemon':4}
```

```python
diccionario_2
```

Ahora las claves son:

```python
diccionario_2.keys()
```

Y se puede obtener un valor de cada clave

```python
diccionario_2['naranja']
```

También podemos añadir otro par clave-valor:

```python
diccionario_2['sandia']=5
```

```python
diccionario_2
```

Nótese que la clave tiene que escribirse exactamente de la misma manera que difinced, si se pone un espacio extra en su nombre o se cambia la letra se obtiene error.

```python
#diccionario_2['   naranja ']
```

```python
#diccionario_2['Naranja ']
```

Además, que la clave es un nombre único, y por lo tanto en el diccionario no pueden existir dos claves con el mismo nombre. Vamos a intentarlo

```python
diccionario_2['manzana']=9
```

```python
diccionario_2 # the value got updated
```

El diccionario se puede crear o actualizar usando loop

```python

keys = ['manzana', 'mandarina', 'naranja']
values = [1, 2, 3]
dictionary_3 = {}
for i in range(len(keys)):
    dictionary_3[keys[i]] = values[i]
print(dictionary_3)
```

```python

keys = ['manzana', 'mandarina', 'naranja']
values = [1, 2, 3]

dictionary_3 = dict(zip(keys,values))
dictionary_3
```

#### Existe otra forma de crear diccionarios más rápida y automática 
Con el mismo ejemplo se definen claves y valores como dos listas separadas:

```python
claves=['fruit_label','fruit_name', 'fruit_subtype','mass', 'width', 'height', 'color_score']
```

Como se tienen siete claves, se necesitan siete valores, que son lista de listas:

```python
valores=[[1,1,1],['apple','apple','apple'],['granny_smith','granny_smith','granny_smith'],[192,180,186], [8.4,7.0,7.4], [7.3,6.8,7.2], [0.55,0.59,0.60]]
```

Y usando función `zip()` y `dict()` se convierten las dos listas a un diccionario:

```python
dictionary_fruits=dict(zip(claves,valores))
```

Que da el mismo resultado que antes:

```python
dictionary_fruits
```

Este diccionario se puede convertir en un dataframe de Pandas (esta vez no se tienen que poner los corchetes extra porque los valores son vectores).

```python
df_fruits = pd.DataFrame(dictionary_fruits)
```

```python
df_fruits.head()
```

#### Actualización del marco de datos Pandas DataFrame

Hay muchas maneras de convertir datos en DataFrames de Pandas. Si se tienen todos los datos listos, se puede crear un diccionario y convertirlo en un marco de datos usando método `pd.DataFrame()`. Pero, a veces, se quiere crear un marco de datos vacío y agregar una nueva fila en un bucle. 

En ese caso, la mejor manera es crear una lista con los nombres de las columnas y crear un marco de datos vacío, por ejemplo:

```python
import pandas as pd
headers=['columna_01','columna_02','columna_03']
df = pd.DataFrame(columns=headers)
```

```python
df
```

Ahora es posible crear una lista con valores y agregarla a la parte inferior del marco de datos vacío utilizando el método `loc` (en el próximo capítulo se profundizará más sobre este)

```python
values=[1,2,3]
df.loc[len(df),:] = values
```

```python
df
```

La lista con valores puede estar cambiando en el bucle que actualiza el marco de datos. 

En otra variación, el marco de datos se puede abrir en el bucle, actualizarlo y guardarlo nuevamente. Esto puede ser útil si existe el miedo de perder datos no guardados en el disco duro.
