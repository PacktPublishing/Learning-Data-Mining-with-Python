### 6. Comprobar la correlación entre variables - selección de características
¿Cómo se eligen qué características en este gran conjunto de datos (columnas) son importantes e influyen en el precio de la casa? Particularmente, ¿cómo saber si dos columnas diferentes no llevan la misma información?

En este conjunto de datos hay más de 20 características diferentes de cada casa, que pueden o no influir en su precio. Una de las columnas se llama `sqft_living`, que se refiere al tamaño de la casa en pies cuadrados. 

Ahora se puede, simplemente, agregar una nueva columna `sqm_living` con un tamaño de la casa en $m^2$, multiplicando la columna `sqft_living` con factor `0.0929`. 

```python
from scipy import constants
df['sqm_living']=df['sqft_living']*constants.foot**2
```

```python
df.head(3)
```

```python
sns.scatterplot(data=df,x='sqft_living',y='sqm_living')
```

La diferencia entre esas dos columnas es solamente un coeficiente, por eso el diagrama de dispersión de los valores de estas columnas muestra una línea recta. Lo que significa que esas dos variables dependen linealmente de cada una o  están correlacionadas linealmente. Como la línea es perfectamente recta, puede decirse que las variables están correlacionadas fuertemente. 

La consecuencia de una fuerte dependencia lineal de estas dos variables es básicamente que `sqm_living` no trae ninguna información nueva sobre el precio de la casa, porque `sqm_living` es simplemente `sqft_living` multiplicada por algún factor. 

Hay otra razón para eliminar características correlacionadas fuertemente: muchos algoritmos de Machine Learning dependen de herramientas de álgebra lineal y resolución de sistemas lineales. Las características correlacionadas pueden crear una matriz indeterminada y singularidades en los algoritmos que las emplean. 

Hay dos pruebas estadísticas que son las más utilizadas para establecer la relación entre variables: ***coeficiente de correlación*** y ***`p-value`*** (valor-p). El coeficiente de correlación es una forma de probar qué tan fuerte es la relación entre dos variables, mientras que el `p-value`dice si el resultado de un experimento es estadísticamente significativo.

#### Coeficiente de correlación

Existen varios tipos de coeficientes de correlación (por ejemplo, Pearson, Kendall, Spearman), pero el más utilizado es el coeficiente de correlación de Pearson, el cual asigna valor -1 a +1. El valor cercano o igual a +1 representa una fuerte relación positiva entre dos características; el valor cercano o igual a -1 representa una fuerte relación negativa entre dos características, mientras que el valor cercano o igual a 0 representa una relación muy débil o nula.

  * +1 Correlación positiva completa   (sería como en el caso de arriba - línea recta con inclinación hacia arriba)
  * +0.8 Fuerte correlación positiva    
  * +0.6 Correlación positiva moderada  
  * 0 Sin correlación  
  * -0.6 Correlación negativa moderada  
  * -0.8 Fuerte correlación negativa  
  * -1 Correlación negativa completa (línea recta con inclinación hacia abajo)  


![](imgs/correlation_coefficients.png)

Cuando las variables están altamente correlacionadas pueden introducir inestabilidad en el modelo. Por eso, es recomendable remover las variables que muestran valor de correlación más alto que abs(0.9).

También, es importante mencionar que `precio` de la casa como variable objetiva tendrá algunas variables linealmente dependientes, pero `NO` se les debe eliminar. Solamente las características de la casa que están linealmente correlacionadas pueden eliminarse.

La correlación de Pearson es el método predeterminado de la función sns.corr() que devuelve una mapa de "calor" de las correlaciones y valores numéricos.

Acá puede verse el mapa de correlación (matriz) del conjunto de datos:

```python
plt.figure(figsize=(20,10))
correlation_values= df.corr()
sns.heatmap(correlation_values,cmap='BrBG',annot=True)
```

Las correlaciones más altas se indican mediante cuadrados verdes más oscuros. Es fácil identificar la línea en el medio, donde cada variable se enfrenta a sí misma.

La matriz de correlación de Pandas puede ayudar a identificar características que tienen una correlación superior a un umbral definido. En este caso se utilizó el 90 % como umbral. Ahora está claro que `sqft_living` y `sqm_living` son solo una combinación lineal con otra, por lo tanto, están correlacionadas al máximo (corr = 1), y una es redundante y debe eliminarse. 
También`sqft_above` y `sqft_living` están demasiado correlacionadas (corr=0.9), por lo que se debe eliminar una. 

Otra vez, las variables linealmente relacionadas con el precio no deben eliminarse porque la columna `precio` es una variable objetiva y es, obviamente, debido a que el `precio` puede depender linealmente de varias características.

### 8. Buscar y eliminar valores atípicos

Un valor atípico se refiere a un punto o conjunto de puntos que son diferentes de otros. A veces pueden ser muy altos o bajos. Generalmente, es una buena idea detectar y eliminar los valores atípicos (obviamente, no es una buena idea eliminar los valores atípicos cuando se estudian anomalías en el sistema) porque son una de las principales razones que dan como resultado un modelo menos preciso. Los valores atípicos son normalmente errores, pero no siempre. Pueden aparecer en el conjunto por diferentes motivos:

* Error humano al ingresar datos manualmente, como un error tipográfico.
* Errores intencionales, como valores atípicos ficticios incluidos en un conjunto de datos para probar los métodos de detección.
* Errores de muestreo que surgen de la extracción o mezcla de datos de fuentes inexactas o diversas.
* Errores de procesamiento de datos que surgen de la manipulación de datos o mutaciones no intencionadas de un conjunto de datos.
* Errores de medición como resultado de un error instrumental.
* Errores experimentales del proceso de extracción de datos o de la planificación o ejecución del experimento.
* Valores atípicos que ocurren "naturalmente" en el conjunto de datos en lugar de ser el resultado de un error enumerado de otra manera. Estos errores naturales se conocen como novedades.

Hay dos técnicas típicas de encontrar y eliminar valores atípicos:

* Método gráfico - visualización
* Método estadístico - IQR (InterQuartile Range) - Rango intercuartil
* Método estadístico - z-score

De los resultados de `df.describe()` quedó claro que el valor de la cantidad de habitaciones para ser igual a 33 parece un valor atípico. Esto se verifica usando scatterplot:

```python
sns.scatterplot(data=df,x='bedrooms',y='price')
```

Desde el gráfico es claro que el valor de `bedrooms=33` es un valor atípico. La casa con 33 cuartos puede ser un error tipográfico o una casa muy grande (hotel); sin embargo, una casa tan grande tendría un precio más alto que una casa típica, pero en este caso no es así.

Ahora se hará uso de una heramienta estadística para buscar los valores atípicos.      
Un diagrama de caja (`boxplot`) es un método para representar gráficamente grupos de datos numéricos a través de sus cuartiles (mencionados en la primera parte de EDA). Los diagramas de caja también pueden tener líneas que se extienden verticalmente desde las cajas (bigotes) que indican variabilidad fuera de los cuartiles superior e inferior; de ahí los términos diagrama de caja y bigotes [18]. Los valores atípicos se pueden trazar como puntos individuales. La definición anterior sugiere que si hay un valor atípico se trazará como punto en boxplot, pero otra población se agrupará y se mostrará como cajas 

```python
sns.boxplot(x=df['bedrooms'])
```

Los valores típicos están adentro de esta barra azul con los "bigotes" y valores atípicos estan afuera de los "bigotes"

Los elementos de diagrama de caja (`boxplot`) tienen la siguiente estructura:

![](imgs/outliers.png)

* Mediana (Q2, percentil 50): el valor medio del conjunto de datos.
* Primer cuartil (Q1, percentil 25): el número medio entre el número más pequeño (no el "mínimo") y la mediana del conjunto de datos. [19] 
* Tercer cuartil (Q3, percentil 75): el valor medio entre la mediana y el valor más alto (no el "máximo") del conjunto de datos.[19]
* Rango intercuartílico (IQR): percentil 25 al 75.
* Bigotes (mostrado en azul)


* Valores atípicos (mostrados como círculos verdes)
  * Más grandes que : Q3 + 1.5*IQR
  * Más pequeños que: Q1 -1.5*IQR

El diagrama de caja utiliza el método IQR para mostrar datos y valores atípicos (forma de los datos), pero para obtener una lista de valores atípicos identificados debe usarse la fórmula matemática y recuperar los datos atípicos.

El rango intercuartílico (IQR), también llamado midspread o mid 50%, es una medida de dispersión estadística, siendo igual a la diferencia entre los percentiles 75 y 25 o entre los cuartiles superior e inferior, IQR = Q3 - Q1. [20]


En otras palabras, el IQR es el primer cuartil restado del tercer cuartil. Estos cuartiles se pueden ver claramente en un diagrama de caja en los datos [21]  


El diagrama de caja es, de hecho, una medida de la dispersión similar a la desviación estándar o varianza, pero es mucho más robusta contra valores atípicos. La relacion entre diagrama de caja y distribucion es presentada en la siguiente imagen:

![](imgs/boxplot.png)

En primer lugar, se observa en el diagrama de densidad de probabilidad que:

* Alrededor del 68,26 % de todos los datos se encuentran dentro de una desviación estándar (<$\sigma$) de la media ($\mu$) teniendo en cuenta ambos lados, la región rosa de la figura.

* Alrededor del 95,44 % de todos los datos se encuentran dentro de dos desviaciones estándar (2$\sigma$) de la media ($\mu$) teniendo en cuenta ambos lados, la región rosa+azul de la figura.

* Alrededor del 99,72 % de todos los datos se encuentran dentro de las tres desviaciones estándar (<3$\sigma$) de la media ($\mu$) teniendo en cuenta ambos lados, la región rosa+azul+verde de la figura.

* Y el 0,28 % restante de todos los datos se encuentra fuera de las tres desviaciones estándar (>3$\sigma$) de la media ($\mu$) teniendo en cuenta ambos lados, la pequeña región roja de la figura.

 Esta parte de los datos se considera como valores atípicos.

Ahora, en el diagrama de caja que corresponde a diagrama de densidad de probabilidad se observa que los cuartiles primero y tercero, Q1 y Q3, se encuentran a -0.675$\sigma$ y +0.675$\sigma$ de la media, respectivamente. Aquí se ubica la relación entre los dos.

El valor $1.5$ de acuerdo con el método IQR corresponde a 2.7$\sigma$ de la media ($\mu$) en cualquier lado y se considerará como atípico. Este rango de decisión es el más cercano a lo que plantea la Distribución Gaussiana, es decir, 3$\sigma$;en otras palabras, esto hace que la regla de decisión se acerque más a lo que Gaussian Distribution considera para la detección de valores atípicos y es exactamente lo que se quería. Para obtener exactamente 3$\sigma$, es necesario tomar la escala = 1.7, pero entonces 1.5 es más “simétrico” que 1.7.

Ahora se calcula Q1, Q3, IQR para cada variable del conjunto:

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(Q3+1.5*IQR)
```

Especificamente, para columna `bedrooms` sería:

```python
IQR['bedrooms']
```

```python
valores_maximos_bedrooms=Q3['bedrooms'] + 1.5 * IQR['bedrooms']
print(valores_maximos_bedrooms)
```

```python
valores_minimos_bedrooms=Q1['bedrooms'] - 1.5 * IQR['bedrooms']
print(valores_minimos_bedrooms)
```

Estos tres valores coresponden al diagrama de caja de la columna `bedroom` que se presentó anteriormente

```python
sns.boxplot(x=df['bedrooms'])
```

* El grosor de la barra azul corresponde a IQR = 1
* El valor mínimo se corresponde con "los bigotes" (línea fina negra de lado izquierdo) = 1.5 
* El valor máximo se corresponde con "los bigotes" (línea fina negra de lado derecho) = 5.5 

```python
sns.countplot(data=df,x='bedrooms')
```

Otro método para eliminar los valores atípicos, aún más intuitivo, es el método llamado `z-score`, el cual indica cuántas desviaciones estándar de distancia están en un punto de datos de la media. Eso ayuda a comprender si un valor de datos es mayor o menor que la media y qué tan lejos está de la media [22] . Si se sigue la regla de los tres sigma - $3 \sigma$, esto solo representa el 0,3% de los datos, lo que parece insignificante, pero puede tener efectos importantes en el modelo. Para calcular el porcentaje de las casas que tienen números atípicos de cuartos (`bedrooms`) con método `z-score`, se usará una combinación de los métodos zscore() y abs() de scipy.

```python
df[(np.abs(stats.zscore(df['bedrooms'])) > 3)].head(10)
```

Ahora, se usa `groupby` para ver cuál es el número de casas con dormitorios que están 3 $\sigma$ fuera de la distribución. 

```python
df_outliers_bedrooms=df[(np.abs(stats.zscore(df['bedrooms'])) > 3)]
df_outliers_bedrooms.groupby('bedrooms').count()['price']
```

Hay en total 61 casas con números atípicos de dormitorios: hay 13 casas sin habitaciones -que puede sonar extraño- y cinco casas que tienen más de 10 habitaciones, incluida una casa que tiene 33 habitaciones (lo que definitivamente parece un error tipográfico).

Acá, se observa cuál es el porcentaje de esas casas atípicas con respecto al número total de casas

```python
print(f'{df_outliers_bedrooms.shape[0]/df.shape[0]*100:.2f}%')
```

Una vez más, para los puntos de demostración se pueden eliminar esos valores atípicos

```python
df=df[(np.abs(stats.zscore(df['bedrooms'])) < 3)]
```

Es importante tener en cuenta que los métodos de boxplot, IQR (InterQuartile Range) y `z-score` para identificar los valores atípicos son solamente herramientas de exploración de datos, que no significa que en todos casos los valores atípicos tengan que ser removidos, por ejemplo, si se quieren analizar las anomalías en el conjunto de datos, obviamente no se querrá eliminar esos valores; por lo tanto, el tratamiento de los valores atípicos depende del objetivo que se desea lograr. La mayoría de los académicos argumentan que, a menos que tenga una razón muy convincente para eliminar una observación particular de su archivo de datos, lidiar con valores atípicos es más problemático de lo que vale. Se podría obtener un conjunto de datos un poco más limpio, pero a expensas de perder puntos de datos quizá significativos.

#### En general...

* Elimine un valor atípico si:
