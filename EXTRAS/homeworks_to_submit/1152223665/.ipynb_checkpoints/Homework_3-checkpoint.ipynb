### Detalles del conjunto de datos
Este conjunto de datos captura los detalles de cómo las emisiones de CO2 de un vehículo pueden variar con las diferentes características. El conjunto de datos se tomó del sitio web oficial de datos abiertos del gobierno de Canadá, esta es una versión compilada y contiene datos de un período de 7 años.

Hay un total de 7385 filas y 12 columnas, además se han utilizado pocas abreviaturas para describir las características.

**1. Model :**
4WD/4X4 = Four-wheel drive, AWD = All-wheel drive, FFV = Flexible-fuel vehicle, SWB = Short wheelbase, LWB = Long wheelbase ,EWB = Extended wheelbase.

**2. Transmission :**
A = automatic, AM = automated manual, AS = automatic with select shift, AV = continuously variable, M = manual, 3-10 = number of gears.

**3. Fuel type :**
X = regular gasoline, Z = premium gasoline, D = diesel, E = ethanol (E85), N = natural gas.

**4. Fuel consumption :**
City and highway fuel consumption ratings are shown in litres per 100 kilometres (L/100 km) - the combined rating (55% city, 45% hwy) is shown in L/100 km and in miles per imperial gallon (mpg).

**5. Estimated tailpipe carbon dioxide emission in grams/km**

# 1. Carga y limpieza de los datos

* a) Completación del conjunto de datos de la emisión de CO2 del automóvil con los datos sobre el origen de las marcas.

```python
data_CO2=pd.read_csv("https://raw.githubusercontent.com/mhemmg/datasets/master/car_emission/cars_CO2_emissions_D.csv")
df=pd.read_csv("https://raw.githubusercontent.com/mhemmg/datasets/master/car_emission/car_brands_countries.csv")

a1=df["Make"].values
a2=data_CO2["Make"].values

cont=np.zeros(len(a2)).astype(str)
countr=np.zeros(len(a2)).astype(str)

for i in range(len(a1)):
    for j in range(len(a2)):
        if a1[i]==a2[j]:
            cont[j]=df["Continent"].values[i]
            countr[j]=df["Country"].values[i]

data_CO2["Country"]=countr
data_CO2["Continent"]=cont

data_CO2
```

```python
data_CO2.info()
```

Las columnas de tipo *Object* son los strings y convertiremos todos en Mayúsculas para homogenizar la información. 

```python
for c in data_CO2.columns:
    if data_CO2[c].dtype == object:
        data_CO2[c] = data_CO2[c].str.upper()

data_CO2
```

```python
data_CO2=data_CO2.replace("?",np.nan) #reemplaza los simbolos "?" por valores nan
data_CO2=data_CO2.replace(r"^\s*$",np.nan,regex=True) #RegEx ^\s*$ significa que el string comience con un espacio y termine 
                                                #con otro y ademas entre ellos debe coincidir cero veces o mas el espacio " ".
                                                #esto corresponde a los valores vacíos que hayan en el dataframe.
                                                #estos valores son reemplazados por nan
```

# 2. Adición y cambio de las características

* a) Cambio de los nombres de la gasolina ('Fuel Type') a nombres más significativos (X = gasolina regular, Z = gasolina premium, D = diesel, E = etanol (E85), N = gas natural).

```python
data_CO2["Fuel Type"].unique()
```

```python
data_CO2.replace("REGULAR GASOLINE","X",inplace=True) 
data_CO2.replace("PREMIUM GASOLINE","Z",inplace=True) 
data_CO2.replace("DIESEL","D",inplace=True) 
data_CO2.replace("ETHANOL","E",inplace=True) 
data_CO2.replace("NATURAL GAS","N",inplace=True)
data_CO2
```

* b) Creamos una nueva columna llamada 'Automatic / Manual' donde solo aparecerán valores de cadenas: A y M (extraiga dichos valores de la columna 'Transmission').

```python
data_CO2["Transmission"].unique()
```

```python
def Masc(val):
    new_val="A"
    if val[0]=="A":
        new_val="A"
    elif val[0]=="M":
        new_val="M"
    return new_val

data_CO2["Automatic / Manual"]=data_CO2["Transmission"].apply(Masc)
data_CO2
```

* c) Creamos una nueva columna llamada ' A / M digital' donde solo aparecerán valores numéricos: 1 / 0. Los valores enteros 1 y 0 serán valores correspondientes al carro con transmision automática o manual (tener una columna con el valor digital de la transmisión puede ser útil al realizar la regresión, porque la regresión usando scikit-learn funciona solo con valores numéricos).

```python
def Mask(val):
    new_val=0
    if val=="A":
        new_val=1
    elif val=="M":
        new_val=0
    return new_val

data_CO2["A / M digital"]=data_CO2["Automatic / Manual"].apply(Mask)
data_CO2
```

# 4. Trazado

* a) Se realiza un 'boxplot' (diagrama de caja) comparando la emisión de los carros por continentes: América, Europa, Asia (añadir línea media para comparación).

```python
sns.set(rc={"figure.figsize":(10, 6)})
ax=sns.boxplot(x='Automatic / Manual',y='CO2 Emissions(g/km)',data=data_CO2)
ax.axhline(data_CO2['CO2 Emissions(g/km)'].mean(),color='r',linestyle='dashed',linewidth=2)
```

* c) Realizamos un 'boxplot' (diagrama de caja) comparando la emisión de los carros por tipo de combustible (añadir línea media para comparación).

```python
sns.set(rc={"figure.figsize":(10, 6)})
ax=sns.boxplot(x='Fuel Type',y='CO2 Emissions(g/km)',data=data_CO2)
ax.axhline(data_CO2['CO2 Emissions(g/km)'].mean(),color='r',linestyle='dashed',linewidth=2)
```

* d) Realizamos un 'countplot' de conteo de automóviles con transmisión manual y automática (solo dos grupos).

```python
sns.set(rc={"figure.figsize":(10, 6)}) # tamano del plot
sns.countplot(data=data_CO2, x='Automatic / Manual')
```

* e) Realizamos un 'barplot' de una emisión promedio de CO2 por marca de automóvil (de menor a mayor).

```python
barplots=data_CO2.groupby(['Make']).mean()["CO2 Emissions(g/km)"]
barplots=barplots.sort_values(ascending=True)
barplots=barplots.rename("CO2 Average Emissions(g/km)")

sns.set(rc={"figure.figsize":(10, 8)})
ax=sns.barplot(x=barplots,y=barplots.index)
```

* f) Ordenamos el conjunto de datos por marca y emisión de CO2 y guárdelo en el archivo CSV (preferiblemente sin guardar el número de índice).
