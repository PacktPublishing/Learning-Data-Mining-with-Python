### Carga y Limpieza de datos.

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/mhemmg/datasets/master/car_emission/cars_CO2_emissions_D.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/mhemmg/datasets/master/car_emission/car_brands_countries.csv')
```

```python
Data = df1.merge(df2, on='Make', how="left")
Data.head()
```

#### E. Llenando celdas vacías.

Las celdas con cadenas vacías `''` o conteniendo el símbolo `'?'` son cambiadas por `NaN`

```python
Data = Data.replace('', np.nan)
Data = Data.replace('?', np.nan)
```

#### F. Conteo y borrado de filas con `NaN`.

```python
#Número de NaN values por columna.
Data.isnull().sum()
```

```python
#Borrado de las filas conteniendo NaN.
Data = Data.dropna()
Data.isnull().sum()
```

#### G. Conteo y borrado de filas duplicadas.

```python
print('Hay {} filas repetidas.'.format(Data.duplicated(keep=False).sum()))
Data = Data.drop_duplicates()
```

### Adición y Cambio de Características. 

#### A. `Fuel Type`.

```python
fuels = {'regular gasoline':'X', 'premium gasoline':'Z', 'diesel':'D', 'ethanol':'E', 'natural gas':'N'}
Data = Data.replace(fuels)
Data.head()
```

#### B. Añadiendo columna `Automatic / Manual`.

```python
Data['Automatic / Manual'] = Data['Transmission'].str.extract(r'([AM])')
Data.head(10)
```

#### C. Añadiendo columna `A / M digital`.

```python
def digital(s):
    d = 0
    if s=='A':
        d = 1
    return d

Data['A / M digital'] = Data['Automatic / Manual'].map(digital)
Data.head(10)
```
