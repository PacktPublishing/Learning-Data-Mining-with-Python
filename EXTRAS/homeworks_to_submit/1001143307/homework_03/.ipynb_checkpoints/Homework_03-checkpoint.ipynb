## 2. AdiciÃ³n y cambio de las caracterÃ­sticas

```python
df['Fuel Type'].unique()
```

```python
# Function to convert the values of the column Fuel Type
# X = gasolina regular, Z = gasolina premium, D = diesel, E = etanol (E85), N = gas natural
def convert(string):
    
    new_string = ''
    
    if string == 'regular gasoline':
        new_string = 'X'
    if string == 'premium gasoline':
        new_string = 'Z'
    if string == 'diesel':
        new_string = 'D'
    if string == 'ethanol':
        new_string = 'E'
    if string == 'natural gas':
        new_string = 'N'
        
    return new_string
```

```python
%time
new_fueltype_column = list(map(convert, df['Fuel Type']))
```

```python
%time
s = df['Fuel Type'].apply(convert)
```

```python
df['Fuel Type'] = df['Fuel Type'].apply(convert)
```

```python
df
```

```python
df['Fuel Type'].unique()
```

```python
df['Transmission'].unique()
```

```python
# Function to convert the strings with 'a' (automatic) with 'A' and Manual with 'M'
def func(word):
    if 'a' in word:
        return 'A'
    else:
        return 'M'
```

```python
df['Automatic / Manual'] = df['Transmission'].apply(func)
```

```python
df
```

```python
df['A/M digital'] = df['Automatic / Manual'].replace(r"A", 1)
df['A/M digital'] = df['A/M digital'].replace(r"M", 0)
```

```python
df
```

## 4. Trazado

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(13,8)})
mean_Co2_emission = df['CO2 Emissions(g/km)'].mean()
sns.boxplot(x='Continent', y='CO2 Emissions(g/km)', data=df)
plt.hlines(mean_Co2_emission, xmin=-0.5, xmax=2.5, color='red', linestyles='--')
```

A travÃ©s del boxplot anterior se puede inferir que el continente con un promedio mÃ¡s alto en la emisiÃ³n de CO2 a causa de los automoviles es `AmÃ©rica`, y que el que menos emisiÃ³n tiene es `Asia`

```python
sns.boxplot(x='Automatic / Manual', y='CO2 Emissions(g/km)', data=df)
plt.hlines(mean_Co2_emission, xmin=-0.5, xmax=1.5, color='red', linestyles='--')
```

A travÃ©s de la grafica se deduce que los carros automaticos son los que mÃ¡s emiten CO2 a comparaciÃ³n con los de transmisiÃ³n  manual.

```python
sns.boxplot(x='Fuel Type', y='CO2 Emissions(g/km)', data=df)
plt.hlines(mean_Co2_emission, xmin=-0.5, xmax=4.5, color='red', linestyles='--')
```

Segun el diagrama el tipo de combustible que mÃ¡s emisiÃ³n de CO2 genera es `E=Etanol`

```python
sns.countplot(x='Automatic / Manual', data=df)
```

Lo dicho de que los automoviles automaticos generan mas CO2 es porque hay mucha mÃ¡s cantidad de ellas en el `DataFrame` que los manuales.

```python
sns.barplot(x='CO2 Emissions(g/km)', y='Make', data=df, order=df.groupby(['Make']).mean()['CO2 Emissions(g/km)'].sort_values(ascending=True).index)
```

```python
# Ordering alphabetical the brands
df = df.sort_values('Make').reset_index()
df.drop(columns='index', inplace=True)
df
```

```python
# Brands ordered alphabetical
sorted_brands = df['Make'].unique()
sorted_brands.sort()

# Empty DataFrame
DF = pd.DataFrame()

# For each brand, ordering from smallest to largest CO2 EMISSION
for mark in sorted_brands:
    DF = pd.concat([DF, df[df['Make']==mark].sort_values('CO2 Emissions(g/km)')])
    
```

```python
DF = DF.reset_index()
DF.drop(columns='index', inplace=True)
```
