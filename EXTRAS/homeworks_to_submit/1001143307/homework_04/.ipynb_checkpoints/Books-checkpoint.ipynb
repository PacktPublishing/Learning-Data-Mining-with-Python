```python
import re
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# Empty list to contain all names of the books in the 50 pages. 
book_names = []
# Empty list to contain all prices of each book
all_prices = []

pages_number = 50
for page in range(1, pages_number + 1):
    # url for each page of the website
    url = f"https://books.toscrape.com/catalogue/page-{page}.html"
    web_pag = requests.get(url)
    # Converting to a BeautifulSoup object
    soup = BeautifulSoup(web_pag.content, 'html.parser')
    
    # Extracting the book's names of each page
    names = [a.attrs.get('title') for a in soup.select('article.product_pod h3 a')]
    # Concating 
    book_names += names
    
    # Extracting the book's prices of each page
    prices = [b.get_text() for b in soup.select('div.product_price p.price_color')]
    # Converting a the prices to floats
    float_prices = [float(re.findall(r"(\d+.\d+)", p)[0]) for p in prices]
    # Concating
    all_prices += float_prices
    
```

```python
dicc = {'Name': book_names, 'Price': all_prices}
df = pd.DataFrame(dicc)
df
```

```python
df.to_csv('Books_data.csv', index=False)
```
