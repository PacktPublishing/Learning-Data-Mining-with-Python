
## 2. AdiciÃ³n y cambio de las caracterÃ­sticas

a) Cambie los nombres de la gasolina ('Fuel Type') a nombres mÃ¡s significativos (X = gasolina regular, Z = gasolina premium, D = diesel, E = etanol (E85), N = gas natural).

```python
def my_func(val):
    new_val="0"
    if val=='regular gasoline':
        new_val="X"
    elif val=='premium gasoline':
        new_val="Z"
    elif val=='diesel':
        new_val="D"
    elif val=='ethanol':
        new_val="E"
    elif val=='natural gas':
        new_val="N"
    return new_val 



```

```python
df['Fuel Type'].unique() # Averiguamos los nombres diferentes en esta columna
```

```python
df['Fuel Type'] = df['Fuel Type'].apply(my_func) #Cambiamos los nombres en la columna Fuel Type
```

```python
df
```

```python
df['Fuel Type'].unique() # Verificamos que los cambios quedaran correctos
```

b) Cree una nueva columna llamada 'Automatic / Manual' donde solo aparecerÃ¡n valores de cadenas: A y M (extraiga dichos valores de la columna 'Transmission')

```python
new_df = df.assign(Automatic_Manual='NAN') #Agregamos la nueva columna
new_df
```

```python
arr=df['Transmission'].unique() 
arr
```

```python
def my_Newfunc(val):
    new_val="0"
    if val[0]=='A':
        new_val="A"
    elif val[0]=='M':
        new_val="M"
    return new_val 
```

```python
new_df['Automatic_Manual'] = new_df['Transmission'].apply(my_Newfunc)

new_df
```

c) Cree una nueva columna llamada ' A / M digital' donde solo aparecerÃ¡n valores numÃ©ricos: 1 / 0. Los valores enteros 1 y 0 serÃ¡n valores correspondientes al carro con transmision automÃ¡tica o manual (tener una columna con el valor digital de la transmisiÃ³n puede ser Ãºtil al realizar la regresiÃ³n, porque la regresiÃ³n usando scikit-learn funciona solo con valores numÃ©ricos).

```python
def my_NewfuncDigital(val):
    new_val=0
    if val=='A':
        new_val=1
    elif val=='M':
        new_val=0
    return new_val 

new_df['A/M digital'] = new_df['Automatic_Manual'].apply(my_NewfuncDigital)

new_df
```
