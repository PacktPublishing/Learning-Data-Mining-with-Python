Con las siguientes lineas se descargará el libro "Adventures of Huckleberry Finn" de la página web

```python
import numpy as np
# Descargamos el libro
from urllib.request import urlopen
contents = urlopen("https://www.gutenberg.org/files/76/76-0.txt").read().decode('utf-8')
print(contents)
```

### Primero vamos a dividir el libro por capitulos
Tomamos como referencia para iniciar el conteo el último renglón antes de CHAPTER I. ya que las cadenas CHAPTER solo aparecen cuando se muestra el contenido y al inicio de cada capítulo. de esta manera con ayuda del metodo find() encontramos el índice de la primera letra en cada uno de los capitulos del libro, analogamente para finalizar el conteo.

```python
indice_de_referencia_inicial = contents.find('Scene: The Mississippi Valley Time: Forty to fifty years ago')

print(f'El indice de referencia del inicio es: {indice_de_referencia_inicial}')

indice_de_referencia_final = contents.find('*** END OF THE PROJECT GUTENBERG EBOOK HUCKLEBERRY FINN ***')

print(f'El indice de referencia del final es: {indice_de_referencia_final}')
```

A continuacion creamos una función que toma números enteros y retorna números romanos, con esto evitamos introducir los números romanos de forma manual

```python
## Creamos una funcion que convierte enteros a romanos de 0 a 50
ListaBase = [(1000, 'M'), (900, 'CM'), (500, 'D'), (400, 'CD'), (100, 'C'), (90, 'XC'),
           (50, 'L'), (40, 'XL'), (10, 'X'), (9, 'IX'), (5, 'V'), (4, 'IV'), (1, 'I')]
def entero_romano(entero):
    romano = ''
    while entero > 0:
        for i, r in ListaBase:
            while entero >= i:
                romano += r
                entero -= i
    return romano
```

Debemos identificar el índice que marca el inicio del siguiente capítulo, de esta manera podremos buscar la frecuencia en las palabras por capítulo

```python
cadena='capitulo'
Dicc_indices={} # Crearemos un diccionario vacío que iremos llenando con ayuda del bucle
for i in range(1,43):
    R=entero_romano(i)
    Capitulo = 'CHAPTER ' + R + '.'
    Dicc_indices[Capitulo]=contents.find(Capitulo, 9509,)
## Agregamos el ultimo capitulo
Dicc_indices['CHAPTER THE LAST']=contents.find('CHAPTER THE LAST', 9509,)
Dicc_indices
```

Ahora encontraremos las frecuencias, para esto se crea una función que recibirá una cadena, el texto donde buscará la cadena y los índices que delimitarán la busqueda por capítulo, esta busqueda la realizaremos mediante el método cout()

```python
def frecuenciaName(Name,texto,indices):
    cadena='capitulo'
    Dicc_Name={'Nombre': Name} # Aqui se guardarán las frecuencias
    for i in range(1,42):
        R=entero_romano(i)
        S=entero_romano(i+1)
        #Busqueda de la cadena por capitulo
        Capitulo = 'CHAPTER ' + R + '.' # Delimita el inicio de la busqueda
        CapituloS = 'CHAPTER ' + S + '.'# Delimita el fin de la busqueda
        Dicc_Name[Capitulo]=texto.count(Name,indices[Capitulo],indices[CapituloS])
        
    #Agregamos los ultimos manualmente
    Dicc_Name['CHAPTER XLII.']=texto.count(Name,indices['CHAPTER XLII.'],indices['CHAPTER THE LAST'])
    Dicc_Name['CHAPTER THE LAST']=texto.count(Name,indices['CHAPTER THE LAST'],584673)
    
    return Dicc_Name
    
frecuenciaTom=frecuenciaName('Tom',contents,Dicc_indices)
frecuenciaJim=frecuenciaName('Jim',contents,Dicc_indices)
frecuenciaHuck=frecuenciaName('Huck',contents,Dicc_indices)
```

### Creando el DataFrame

```python
import pandas as pd

df=pd.DataFrame([frecuenciaTom,frecuenciaJim,frecuenciaHuck])

df
```

```python
df.to_csv("frecuenciasNombres.csv") # Guarda el DataFrame
```

### Frecuencia acumulada

Creamos unas listas donde se guardarán las frecuencias de las cadenas

```python
Lista_Tom=[]
Lista_Jim=[]
Lista_Huck=[]

for i in frecuenciaTom:
    Lista_Tom.append(frecuenciaTom[i])
    Lista_Jim.append(frecuenciaJim[i])
    Lista_Huck.append(frecuenciaHuck[i])
    
del Lista_Tom[0]
del Lista_Jim[0]
del Lista_Huck[0]
```

```python
# Tomamos la ayuda sugerida para calcular la suma acumulada
tom_capitulos_numpy=np.array(Lista_Tom)
tom_capitulos_cumsum=np.cumsum(tom_capitulos_numpy)

print(Lista_Tom)
print(tom_capitulos_cumsum)
```

```python
import matplotlib.pyplot as plot

#edades = [12, 15, 13, 12, 18, 20, 19, 20, 13, 12, 13, 17, 15, 16, 13, 14, 13, 17, 19]

plot.figure(figsize=(20,8))
intervalos = range(min(Lista_Tom), max(Lista_Tom) + 2) #calculamos los extremos de los intervalos
plot.hist(x=Lista_Tom, bins=intervalos, color='red', rwidth=0.85)
plot.title('Histograma de frecuencias',fontsize=35)
plot.xlabel('Tom',fontsize=30)
plot.ylabel('Frecuencia',fontsize=30)
plot.xticks(intervalos)
#plot.grid()
plot.show() #dibujamos el histograma
```

```python
Cap=np.arange(1,44)
plot.figure(figsize=(20,8))
plot.scatter(Cap,tom_capitulos_cumsum,s=100)
plot.title('Suma acumulada',fontsize=35)
plot.xlabel('Capitulo',fontsize=30)
plot.ylabel('Ocurrencia acumulada',fontsize=30)
plot.grid()
plot.show()
```
