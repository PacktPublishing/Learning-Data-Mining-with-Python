```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

#Exercise 1
Using web-scrapping technique, save to a csv file all the names and prices of the books from the following website (1-50):
https://books.toscrape.com/

```python
page  = requests.get("https://books.toscrape.com/")
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python

price=[]
names=[]

base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
for j in range (1,4):
    scrape_url = base_url.format(j)
    page  = requests.get(scrape_url)
    soup = BeautifulSoup(page.content, 'html.parser')
    price = price+soup.select('.price_color')
    names = names+soup.select('.thumbnail')

Names=[]
D=[]
Prices=[]
for i in range(len(names)):
  D.append(re.findall(r'class.+',str(names[i])))
  names[i]=re.sub(D[i][0],'',str(names[i]))
  Names.append(re.findall(r'"(.+)"',names[i])[0])
  Prices.append(".".join(re.findall(r'(\d+)',str(price[i]))))

```

```python
NAMES=Names[0:50]
PRICES = Prices[0:50]
```

```python
df = pd.DataFrame(list(zip(NAMES,PRICES)), columns = ['Name','Price'])
print(df)
```

```python
df.to_csv('Books_names&prices.csv',index=False)
```
