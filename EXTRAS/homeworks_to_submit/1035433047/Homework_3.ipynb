```python
df['Season'].unique()
```

```python
s_p=df[df['Season']=='Summer']['Year'].value_counts().sort_index().values
s_p_y=df[df['Season']=='Summer']['Year'].value_counts().sort_index().index
w_p=df[df['Season']=='Winter']['Year'].value_counts().sort_index().values
w_p_y=df[df['Season']=='Winter']['Year'].value_counts().sort_index().index
plt.figure(figsize=(13,8))
plt.plot(s_p_y,s_p,'--or',label=('Summer'))
plt.plot(w_p_y,w_p,'--ob',label=('Winter'))
plt.xticks(years, rotation=90)
plt.xlabel('Years',fontsize=14)
plt.ylabel('Players number',fontsize=14)
plt.title('Olympic games',fontsize=14)
plt.grid()
plt.legend(fontsize=12)
plt.show()
```

```python
df['Sex'].unique()
```

```python
s_p_m=df[(df['Season']=='Summer')&(df['Sex']=='M')]['Year'].value_counts().sort_index().values
s_p_f=df[(df['Season']=='Summer')&(df['Sex']=='F')]['Year'].value_counts().sort_index().values
s_p_f_y=df[(df['Season']=='Summer')&(df['Sex']=='F')]['Year'].value_counts().sort_index().index
w_p_m=df[(df['Season']=='Winter')&(df['Sex']=='M')]['Year'].value_counts().sort_index().values
w_p_f=df[(df['Season']=='Winter')&(df['Sex']=='F')]['Year'].value_counts().sort_index().values
plt.figure(figsize=(13,8))
plt.plot(s_p_y,s_p_m,'--or',label=('Summer_p_m'))
plt.plot(s_p_f_y,s_p_f,'--ob',label=('Summer_p_f'))
plt.plot(w_p_y,w_p_m,'-s',color='maroon',label=('Winter_p_m'))
plt.plot(w_p_y,w_p_f,'-s',color='navy',label=('Winter_p_f'))
plt.xticks(years, rotation=90)
plt.yticks(np.linspace(0,9000,7))
plt.xlabel('Years',fontsize=14)
plt.ylabel('Players number',fontsize=14)
plt.title('Olympic games',fontsize=14)
plt.grid()
plt.legend(fontsize=12)
plt.show()
```

```python
age_s_m=[np.round(df[(df['Year']==i)&(df['Season']=='Summer')&(df['Sex']=='M')]['Age'].dropna().values.mean(),0) for i in s_p_y] 
age_s_f=[np.round(df[(df['Year']==i)&(df['Season']=='Summer')&(df['Sex']=='F')]['Age'].dropna().values.mean(),0) for i in s_p_f_y]
age_w_m=[np.round(df[(df['Year']==i)&(df['Season']=='Winter')&(df['Sex']=='M')]['Age'].dropna().values.mean(),0) for i in w_p_y]
age_w_f=[np.round(df[(df['Year']==i)&(df['Season']=='Winter')&(df['Sex']=='F')]['Age'].dropna().values.mean(),0) for i in w_p_y]
```

```python
plt.figure(figsize=(12,8))
plt.plot(s_p_y,age_s_m,'--or',label='Summer_m_p')
plt.plot(s_p_f_y,age_s_f,'--ob',label='Summer_f_p')
plt.plot(w_p_y,age_w_m,'-s',color='maroon',label='Winter_m_p')
plt.plot(w_p_y,age_w_f,'-s',color='navy',label='Winter_f_p')
plt.xticks(s_p_y,rotation=90)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Mean age', fontsize=12)
plt.grid()
plt.legend(fontsize=14)
plt.show()
```
