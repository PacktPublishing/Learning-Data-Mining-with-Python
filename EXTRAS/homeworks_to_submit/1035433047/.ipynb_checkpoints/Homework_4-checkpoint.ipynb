```python
import re
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
for i in range(1,51):
  if i==1:
    url = 'https://books.toscrape.com/'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    NewBookNames=[]
    for j in range(len(soup.select('img'))):
      NewBookNames.append(re.findall(r'"(.*)"\sclass',str(soup.select('img')[j]))[0])
    PriceBook=[]
    for j in range(len(soup.find_all(class_='price_color'))):
      PriceBook.append(re.findall(r'[£]{1}\d+.\d+',str(soup.find_all(class_='price_color')[j]))[0])
    page_1=np.hstack((np.array([np.array(NewBookNames)]).T,np.array([np.array(PriceBook)]).T))
  else:
    test=['https://books.toscrape.com/catalogue/page-',str(i),'.html']
    url="".join(test)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    NewBookNames=[]
    for j in range(len(soup.select('img'))):
      try:
        NewBookNames.append(re.findall(r'"(.*)"\sclass',str(soup.select('img')[j]))[0])
      except:
        NewBookNames.append(str(soup.select('img')[j]).split('"')[1])
    PriceBook=[]
    for j in range(len(soup.find_all(class_='price_color'))):
      PriceBook.append(re.findall(r'[£]{1}\d+.\d+',str(soup.find_all(class_='price_color')[j]))[0])
    New_page=np.hstack((np.array([np.array(NewBookNames)]).T,np.array([np.array(PriceBook)]).T))
    page_1=np.hstack((page_1,New_page))
```

```python
Columns=[]
for i in range(1,51):
  Columns.append("".join(['BooksPage',str(i)]))
  Columns.append("".join(['PricesBooksPage',str(i)]))
```

```python
df_NamesAndPrices=pd.DataFrame(page_1,columns=Columns,index=range(1,21))
```
