```python
import requests
url = 'https://www.gutenberg.org/files/76/76-0.txt'
f = requests.get(url)
```

```python
data=f.text
```

```python
data[0:1000]
```

a) Analice el libro contando la aparición de tres nombres de personajes principales: Jim, Tom, Huck a través de los capítulos del libro.

```python
dataSeparated=data.split('CHAPTER')
dataSeparated
```

```python
chapters=[dataSeparated[i] for i in range(44,87)]
```

```python
chapters
```

```python

count_characters={}
for i in range(len(chapters)):

  numberCharacterPerChapter={}
  numberCharacterPerChapter['Jim']=chapters[i].count('Jim')
  numberCharacterPerChapter['Tom']=chapters[i].count('Tom')
  numberCharacterPerChapter['Huck']=chapters[i].count('Huck')
  
  count_characters['chapter'+str(i)]=numberCharacterPerChapter
```

```python
count_characters
```

b) Guarde sus resultados en un formato Pandas DataFrame para que se vea de la siguiente manera:

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(count_characters)
```

d) Importe el archivo CSV guardado usando Pandas y haga un gráfico de la ocurrencia acumulada (cumsum) de los caracteres.


```python
df=pd.read_csv(path+"count_chapters.csv")
```

```python
cumSum=df.cumsum(axis=1)
cumSum.head()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(16,6))
plt.plot(cumSum.loc['Jim'].to_numpy() ,label='Jim')
plt.plot(cumSum.loc['Tom'].to_numpy() ,label='Tom')
plt.plot(cumSum.loc['Huck'].to_numpy() ,label='Huck')
plt.grid()
plt.ylabel("Cumulative sum of appearances")
plt.xlabel("Chapter number")
plt.xticks(np.arange(0,42+1,2))
plt.legend()
plt.show()
plt.close()
```

- Se observa que Jim a lo largo de la historia es el personaje más nombrado, de tal manera que este podría ser el personaje principal. 

- Tom a lo largo de los primeros 3/4 de capítulos posee una suma cumulativa constante, sin embargo, este aumenta al final, de tal manera que se convierte en un personaje protagónico.

- Huck es de los tres el personaje menos nombrado, de tal manera que este podria ser el personaje menos importante
