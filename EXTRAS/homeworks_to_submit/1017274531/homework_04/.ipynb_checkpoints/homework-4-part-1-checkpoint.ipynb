```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://books.toscrape.com/")
soup=BeautifulSoup(page.content, 'html.parser')

print(soup.prettify())
```

```python
def scrape_page(page_url):
    
    '''
    function that return a dataframe with the 
    the name and the price of a book given the page_url
    '''
    
    page=requests.get(page_url)
    soup=BeautifulSoup(page.content, 'html.parser')
    
    names = [img.attrs.get('alt') for img in soup.select('article.product_pod a img')]
    prices=[float(price.get_text()[1:]) 
            for price in soup.select('article.product_pod p.price_color')]
    
    df_temp=pd.DataFrame({'name':names, 'price':prices})
    return df_temp
```

Scraping the first page

```python
df=scrape_page("https://books.toscrape.com/")
```

Scraping the other 50 pages

```python
for i in range(2,50+1):
    page_url=f'https://books.toscrape.com/catalogue/page-{i}.html'
    df_temp=scrape_page(page_url)
    df=pd.concat([df,df_temp])
```
