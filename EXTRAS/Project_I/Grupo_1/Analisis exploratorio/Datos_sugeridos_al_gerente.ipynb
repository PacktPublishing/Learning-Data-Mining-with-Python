```python
data = pd.read_csv("CPT_PREDICTIVE_PPM.csv",encoding='windows-1252')
data.head(3)
```

```python
# Columnas que no aportan para la prediccion
#work order number
#site number
#wo current status
col = []
for columns in data.columns:
    temp = columns.replace(" ","_")
    temp = temp.replace("-","_")
    temp = temp.replace(".","")
    temp = temp.lower()
    col.append(temp)
data.columns = col
data.head()
```

¿Por qué hay datos incompletos en date_completed?

Tres posibilidades

* La nevera ya no pudo ser arreglada
* La nevera se arregló y quien completa los datos olvidó registrarlo
* La alarma salta pero no hay un fallo en la nevera sino en la alarma

Se le sugiere al administrador recolectar datos de fallos en la alarma y registrarlos, con esto las predicciones para la cantidad de refrigeradores que se dañarán en el futuro podría mejorar.

```python
#  numero a alarmas por tienda 
cout = data["site_number"].value_counts()
cout
```

```python
dat = pd.DataFrame(cout.reset_index())
dat.columns = ["site_number","cout"]
dat.head()
```

```python
# las tiendas con mas alarmas
plt.barh(dat["site_number"].head(20).to_numpy(),dat["cout"].head(20).to_numpy())
plt.show()
```

```python
# que sucede si miramos el tamaño de las tiendas
tam = data.loc[:, ['site_number', 'sales_sq_ft'] ]
```

```python
comp = pd.merge(dat, tam, on=[ "site_number"])
comp = comp.drop_duplicates()
comp.head()
```

```python
# normalicemos el numero de alamas por el tamaño de la tienda
plt.barh(comp["site_number"].head(20).to_numpy(),comp["cout"].head(20).to_numpy()/comp["sales_sq_ft"].head(20).to_numpy())
plt.show()
```

vemos que en el mismo orden de la tienda con mas alarmas a la menos (20 primeras) (de abajo hacia arriba) no se tiene la misma correspondencia si se normaliza con el tamaño de las tiendas
