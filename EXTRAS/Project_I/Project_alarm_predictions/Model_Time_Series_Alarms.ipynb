```python
import numpy as np
import pandas as pd
from datetime import *
```

```python
TODO_3meses = pd.read_csv("TODO_3meses.csv")
TODO_3meses.drop(columns=["Unnamed: 0"],inplace=True)
TODO_3meses
```

```python
tiendas = TODO_3meses.Tienda.unique()
to_numerical = np.arange(1, len(tiendas)+1)

dcc = dict(zip(tiendas,to_numerical))

TODO_3meses["Tienda_Num"] = TODO_3meses["Tienda"].apply(lambda x: dcc[x])
TODO_3meses.drop(columns=["Tienda"],inplace=True)
TODO_3meses
```

```python
TODO_3meses.sort_values(by=["Tienda_Num", "semanas"], inplace=True, ignore_index=True)
```

```python
TODO_3meses
```

```python
TODO_3meses.groupby(by=["semanas"]).count()
```

#Split

```python
split_point = int(200*0.8)
data_train = TODO_3meses[TODO_3meses["semanas"] < split_point].copy()
data_val = TODO_3meses[TODO_3meses["semanas"] >= split_point].copy()

X_train, y_train = data_train[[i for i in data_train.columns if i!="alarmas_por_semana"]].values, data_train.alarmas_por_semana.values
X_test, y_test = data_val[[i for i in data_train.columns if i!="alarmas_por_semana"]].values, data_val.alarmas_por_semana.values
```

#train a predictive model

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
#from fbprophet import Prophet
```

```python
#model1 = XGBRegressor(n_estimators=1000, max_depth=4, eta=0.1, subsample=0.7, colsample_bytree=0.8)
model2 = RandomForestRegressor(random_state=0)
#model3 = GradientBoostingRegressor(n_estimators=1000, random_state=0)
parameters = {'criterion':('squared_error', 'absolute_error', 'poisson'), 'max_depth': np.arange(1,16)}
model4 = GridSearchCV(model2, parameters)

#model1.fit(X_train, y_train)
#model2.fit(X_train, y_train)
#model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
model4.best_params_
```

```python
#print ("Model1 train accuracy %.2f"%model1.score(X_train,y_train))
#print ("Model1 test accuracy  %.2f"%model1.score(X_test,y_test))
print ("Model2 train accuracy %.2f"%model2.score(X_train,y_train))
print ("Model2 test accuracy  %.2f"%model2.score(X_test,y_test))
#print ("Model3 train accuracy %.2f"%model3.score(X_train,y_train))
#print ("Model3 test accuracy  %.2f"%model3.score(X_test,y_test))
print ("Model4 train accuracy %.2f"%model4.score(X_train,y_train))
print ("Model4 test accuracy  %.2f"%model4.score(X_test,y_test))
```
